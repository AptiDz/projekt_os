### Se även till att plotta några intressanta features, exempelvis:

g. diagram över könsfördelningen

```python
import plotly.express as px

fig = px.pie(df, names = "Sex")
fig
```

h. diagram över topp 10 länder som tagit flest medaljer

```python
medals = df[df["Medal"].notna()]


largest_countries = medals.groupby("NOC").size().nlargest(10).reset_index(name="Medal Count")
largest_countries.style.hide(axis="index")
```

```python
import matplotlib.pyplot as plt

largest_countries.plot(
    x="NOC", 
    y="Medal Count", 
    kind="bar", 
    color="yellow", 
    title="Top 10 Countries Wins in Medals",
    legend=False
)
plt.xlabel("Countries")
plt.ylabel("Medal Count")
plt.show()
```
